```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
occupancy_grid = []
row = []
with open('/tmp/memory_dump.txt') as f:
    for i, line in enumerate(f):
        if i % 256 == 0 and i != 0:
            occupancy_grid.append(row)
            row = []
        row.append(int(line, base=2))

occupancy_np = np.array(occupancy_grid)
occupancy_np
```

```python
# Create a black and white plot of the occupancy grid
plt.imshow(occupancy_np, cmap='gray')
```
